## Bernoulli
- https://en.wikipedia.org/wiki/Bernoulli_distribution

If $X$ is a random variable with this distribution, 
then:

$ X \in \{0,1\} $

$ \begin{cases}
  Pr("X=1") = p \\
  Pr("X=0") = q 
  \end{cases}$



````python
import matplotlib.pyplot as pl

x= [0, 1]

p= .6
q= 1-p # .4
pmf= [q, p]

pl.stem(x,pmf, use_line_collection=True)

pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')
````

````python
def ryBernoulli(p=.5, linefmt=None, markerfmt=None, basefmt=None):
    x= [0, 1]
    #p= .6
    q= 1-p # .4
    pmf= [q, p]
    pl.stem(x, pmf, 
            use_line_collection=True, 
            linefmt=linefmt,
            markerfmt= markerfmt,
            basefmt= basefmt)
    pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')


pl.figure()
ryBernoulli(.2, 'C0--', 'C0o')

pl.figure()
ryBernoulli(.5, 'C1-.', 'C1x')

pl.figure()
ryBernoulli(.8, 'C2:' , 'C2s')

# 通通畫在 1張 圖中
pl.figure()
ryBernoulli(.2, 'C0--', 'C0o')
ryBernoulli(.5, 'C1-.', 'C1x')
ryBernoulli(.8, 'C2:' , 'C2s')
````

## Binomial
- https://en.wikipedia.org/wiki/Binomial_distribution

- the ```Binomial distribution``` 
- with parameters $n$ and $p$ 
- is the discrete probability distribution 
- of the ```number of successes``` 
- in a sequence of $n$ independent experiments, 
- 
- each asking a yes–no question, 
- and each with its own boolean-valued outcome:
- 
- success/yes/true/ one/ 真/陽/勝/正 (with probability $p$) 
- or 
- failure/no/ false/zero/假/陰/敗/負 (with probability $q = 1 − p$).



----
Notation	

$ B(n,p)$

----
Parameters	

$ n\in \{0,1,2,\ldots \}$
– number of trials

$ p\in [0,1]$
– success probability for each trial

$ q=1-p$

----
Support	

$ k\in \{0,1,\ldots ,n\}$
– number of successes

----
pmf	

$ {\binom {n}{k}}p^{k}q^{n-k}$

----
CDF	

$ I_{q}(n-k,1+k)$

----
Mean	

$ np$

----
Median	

$ \lfloor np\rfloor$
or 
$ \lceil np\rceil $

----
Mode	

$ \lfloor (n+1)p\rfloor $  
or 
$ \lceil (n+1)p\rceil -1$

----
Variance	

$ npq$

----
Skewness	

$ {\frac {q-p}{\sqrt {npq}}}$

----
Ex. kurtosis	

$ {\frac {1-6pq}{npq}}$



````python
import matplotlib.pyplot as pl
import scipy.stats as st
import numpy as np

n= 10
p= .5
q= 1-p 
x= np.arange(0,n+1)


pmf= st.binom.pmf(x, n= n, p= p)

pl.stem(x,pmf, use_line_collection=True)

pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')
````

````python
def ryBinom(p=.5, n=10):
    #n= 10
    #p= .5
    #q= 1-p 
    x= np.arange(0,n+1)
    pmf= st.binom.pmf(x, n= n, p= p)
    pl.stem(x,pmf, use_line_collection=True)
    pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')

pl.figure()
ryBinom(.2)

pl.figure()
ryBinom(.5)

pl.figure()
ryBinom(.8)
````

## Negative Binomial
- https://en.wikipedia.org/wiki/Negative_binomial_distribution
 

the number of successes 
in a sequence of independent and identically distributed Bernoulli trials before a specified (non-random) number of failures (denoted r) occurs.

@ Wikipidia
----

在失敗 r 次之前，成功的次數 X=k
p: 單次成功機率

The probability mass function (```pmf```) of the ```Negative binomial``` distribution is

$ 
pmf(k;r,p)\\
\equiv \Pr("X=k")\\
={\binom {k+r-1}{k}}(1-p)^{r}p^{k}
$

@ scipy.stats
----

st.nbinom?

失敗的次數設為隨機變數 X

直到成功 n 次之時， 

p: 單次成功機率

Prob("X=k") = st.binom(k,n,p)

The probability mass function 

of ```the number of failures``` $k$ 

$
   f(k) = \binom{k+n-1}{n-1} p^n (1-p)^k \\
   = \binom{k+n-1}{k} p^n q^k
$

for $k \ge 0$.

`n` is the number of successes, 
whereas `p` is the probability of a single success.

````python
N= 30
r= 10  # 失敗次數
p= .5  # 單次成功機率
q= 1-p
x= np.arange(0, N+1) # 成功次數

pmf= st.nbinom.pmf(k= x, n= r, p= q)

#from scipy.special import comb
#pmf= comb(k+r-1, k) * (1-p)**r *p**k

pl.stem(pmf, use_line_collection=True)
pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')

````

## Geometric
- https://en.wikipedia.org/wiki/Geometric_distribution

至少有 3 種講法：

(1) 在 1次失敗前，成功 X 次: $ f(x) = p^x * q, x \in \{0,1,2,...\}$

(2) 在 1次成功前，失敗 X 次: $ f(x) = q^x * p, x \in \{0,1,2,...\}$

(3) 在 1次成功前，失敗 X-1 次，成敗總共 X 次: $ f(x) = q^{x-1} * p, x \in \{1,2,3,...\}$

$p$ : 單次成功機率

$q= 1-p$


st.geom?

(3) 在 1次成功前，失敗 X-1 次，成敗總共 X 次: $ f(x) = q^{x-1} * p, x \in \{1,2,3,...\}$


Notes
-----

The probability mass function for `geom` is:


$
 f(k) = (1-p)^{k-1} p
$

$k \ge 1 $


````python
N= 10

p= .1

x= np.arange(0, N+1) # 【成+敗】之次數，從 1 起跳。

pmf= st.geom.pmf(k= x, p= p)

pl.stem(pmf, use_line_collection=True)
pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')
````

## Hypergeometric
- https://en.wikipedia.org/wiki/Hypergeometric_distribution

 - $X=k$ successes in $n$ draws, 
 -  ```without replacement```, 
 - from a finite population of size $N$ 
 - that contains exactly $K$ objects with that feature (success)
 
 

----
Parameters	

$ {\begin{aligned}
    N&\in \{0,1,2,\dots   \} \\
    K&\in \{0,1,2,\dots ,N\} \\
    n&\in \{0,1,2,\dots ,N\}
   \end{aligned}
   }$

----
Support

$ {k\,\in \,\{\max {(0,\,n+K-N)},\,\dots ,\,\min {(n,\,K)}\}}$

----
pmf	

$   {
     {K \choose k} {{N-K}\choose{n-k}}
    \over
     {N \choose n}
    }$

----
Mean	

$ n {K \over N}$

----
Variance	

$ n {K \over N} {(N-K)\over N} {(N-n) \over (N-1)}$


````python
st.hypergeom?
````

The hypergeometric distribution models drawing objects from a bin.
- `M` is the total number of objects, 
- `n` is total number of Type I objects.
- The random variate ```X=k``` represents the number of Type I objects 
- in `N` drawn without replacement from the total population.

Notes
-----
The symbols used to denote the shape parameters (`M`, `n`, and `N`) are not
universally accepted.   

$
p(k, M, n, N)\\
= \frac{\binom{n}{k} \binom{M - n}{N - k}}
       {\binom{M}{N}}
$

$k \in [\max(0, N + n - M ), \min(n, N)] $



*** | @scipy | @wikipedia
----| ---- | ----
popSize=     |M | N
sucSize=     |n | K
sampSize=    |N | n
sucNumber (r.v.) |X=k | X=k

````python
M=  20  # popSize

n=  10  # sucSize

N=  10   # sampSize

x= np.arange(0, N+1) # 成功之次數。

pmf= st.hypergeom.pmf(k= x, M= M, n= n, N= N)

pl.stem(x, pmf, use_line_collection=True)
pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')
````

## Poisson
- https://en.wikipedia.org/wiki/Poisson_distribution

- A discrete random variable X  
- is said to have a Poisson distribution 
- with parameter λ > 0, 
- for k = 0, 1, 2, ..., 
- the probability mass function of X  is given by

$
f(k;\lambda )\\
=\Pr("X=k")\\
={{λ^k} \over {k!}}   {e^{-λ}}
$

The positive real number $λ$ is equal to the ```expected value``` of $X$ and also to its ```variance```

$
\lambda \\
=\operatorname {E} (X)\\
=\operatorname {Var} (X).
$


````python
st.poisson?
````

 Notes
-----
$
    f(k) = e^{-\mu} \frac{\mu^k}{k!}
$
$k \ge 0$

````python
μ = 10  # 【平均】成功次數

x= np.arange(0, 3*μ+1) # 成功之次數。

pmf= st.poisson.pmf(k= x, mu= μ)

pl.stem(x,pmf, use_line_collection=True)
pl.xlabel('x'); pl.ylabel('Pr'); pl.grid('on')
````

## Multinomial
- https://en.wikipedia.org/wiki/Multinomial_distribution

 

----

一顆骰子有($k$ = 6 面)，

把它丟擲 ($n$ = 10)次，

每面的數字(1,2,3,4,5,6) 
各出現 $X_1, X_2, X_3, X_4, X_5, X_6$ 次， 

$X_1 + X_2+ X_3+ X_4+ X_5+ X_6 = 10 $

這 6 個 隨機變數 $X_1, X_2, X_3, X_4, X_5, X_6$ 的 ```【聯合】機率質量函數 (joint pmf)```

已知每面的數字出現的機率分別為

$p_1, p_2, p_3, p_4, p_5, p_6$

且
$p_1+ p_2+ p_3+ p_4+ p_5+ p_6 = 1$

The probability mass function of this multinomial distribution is:

is:

$ 
f(x_1,\ldots ,x_k;n,p_1,\ldots ,p_k)
$

$
=\Pr("X_1=x_1" {\text{ and }} "X_2=x_2" \dots {\text{ and }} "X_k=x_k")
$

$
=
\begin{cases}
\binom{n}{x_1,x_2,\dots, x_k} 
p_1^{x_1} \cdot p_2^{x_2} \cdots p_k^{x_k},
 \text{when }  \sum_{i=1}^{k}x_i=n \\\\
 0, \quad {\text{otherwise}}
\end{cases}
$

----
Mean	

$
\operatorname {E} (X_{i})=np_{i}
$

----
Variance	

$
\operatorname {Var} (X_{i})=np_{i}(1-p_{i})
$

$
\operatorname {Cov} (X_{i},X_{j})=-np_{i}p_{j}
$



````python
#
# generate all possible X=[x1,x2,x3,x4,x5,x6]
#
n= 10
X= [(x1,x2,x3,x4,x5,x6) 
 for x1 in range(n+1) 
 for x2 in range(n+1) 
 for x3 in range(n+1)
 for x4 in range(n+1)
 for x5 in range(n+1)
 for x6 in range(n+1)
 if x1+x2+x3+x4+x5+x6==n]
X= np.array(X)
X.shape, X

````

## Negative Binomial r.v.

- If repeated independent trials can result in 
- a ```success``` with probability $p$ 
- and
- a ```failure``` with probability $q = 1 − p$, 
- 
- then the probability distribution of the
- random variable $X$, the number of the trial 
- on which the $k^{th}$ success occurs, is

$
b^{*}(x; k, p) =
\binom{x − 1}{k − 1}
p^k q^{x−k}, \text{where } x \ge k
$


----

【rySol】
====

為了把 Negative Binomial 和 Binomial 放在一起講，以利比較，我重新修正一下符號的使用。

令 N, K, R  分別為
- 獨立試驗(experiment)的次數 N、
- 成功(success)的次數 K、
- 失敗(failure)的次數 R。

簡稱為 N試、K成、R敗。

在每個單次試驗中，
- 成功機率為 p, 
- 失敗機率為 q.

2 個恆等式先寫下來：
- K+R=N
- p+q=1

    - 大寫的 N,K,R 代表隨機變數 (random variables)
    - 小寫的 n,k,r 代表特定數值 (deterministic values)

詳細考慮以下 3 個語句：
- $S_1$. "在 n 次試驗中，成功 k 次"
- $S_2$. "在 第n次那1次試驗中成功 1 次，```且``` 前 (n-1)次試驗中 成功 (k-1)次。
- $S_3$. "在 第n次那1次試驗中成功 0 次，```且``` 前 (n-1)次試驗中 成功 (k)次。

我們發現：
$S_1 = S_2 \cup S_3$，
而且 
$S_2 \cap S_3 =\phi$。

因此：

$
P(S_1) = P(S_2 \cup S_3) = P(S_2) + P(S_3)
$

令 $f(n,k)$ 代表 $P("在 n 次試驗中，成功 k 次")$

則
$
P(S_1) = f(n,k)                 \\
P(S_2) = f(1,1) \cdot f(n-1,k-1)\\
P(S_3) = f(1,0) \cdot f(n-1,k)  \\ 
$

$
P(S_1) = f(n,k) = \binom{n}{k} p^{k} q^{n-k}    \\
\\
P(S_2) = f(1,1) \cdot f(n-1,k-1)
=\binom{1}{1} p^{1} q^{0} \binom{n-1}{k-1} p^{k-1} q^{n-k}

\\
P(S_3) = f(1,0) \cdot f(n-1,k)  
=\binom{1}{0} p^{0} q^{1} \binom{n-1}{k} p^{k} q^{n-k-1} 
=\binom{n-1}{k} p^{k} q^{n-k}
\\
$

$
\binom{n}{k} p^{k} q^{n-k}
== \binom{n-1}{k-1} p^{k} q^{n-k} 
+ \binom{n-1}{k} p^{k} q^{n-k}
$

$
\binom{n}{k}
== \binom{n-1}{k-1} 
+ \binom{n-1}{k}

## Pascal Random Variable

## The Negative Binomial Random Variable 

https://www.sciencedirect.com/topics/mathematics/pascal-random-variable

The Pascal random variable $X_k$ is an extension of the geometric random variable. 

$X_k$ describes ```the number of trials until the kth success```, 

which is why it is sometimes called the kth-order ```interarrival time``` for a Bernoulli process. 

The ```Pascal distribution``` is also called the ```Negative Binomial distribution```.


Let $X_k$ be a kth-order Pascal random variable. 
Then its PMF is given by

$
p_{X_k}(x) = \binom{x−1}{k−1} p^k (1−p)^{x − k},\\
k = 1 , 2 , … ; \\
x = k , k + 1 , …
$

$X_k$ is essentially the sum of k independent ```geometric``` random variables

----
Let $R_k$ = $X_k - k$  代表 直到 【k 勝】時，【敗之數】。

Then its PMF is given by

$
p_{R_k}(r) \equiv Prob("R_k=r")\\
= Prob("X_k - k = r")\\
= Prob("X_k = k + r")\\
= p_{X_k}( x= k+r)\\
= \binom{k+r−1}{k−1} p^k (1−p)^{k+r − k},\\
= \binom{k+r−1}{k−1} p^k q^{r},\\
= \binom{k+r−1}{r} p^k q^{r},\\
= \binom{-k}{r}(-1)^r p^k q^{r},\\
= \binom{-k}{r} p^k (-q)^{r},\\
$

#### 特別注意這一項：
$
\binom{-k}{r} (-q)^{r} \\
它是\\ 
(1-q)^{-k} =  \sum_{r=0}^{-\infty} \binom{-k}{r}  (-q)^{r}\\
之中的一般項，其中：\\
\binom{-k}{r} 
\equiv 
\frac {(-k) (-k-1) (-k-2) ... (-k-(r-1)) }
      { (r) (r-1) (r-2) .... (1)}\\
=\frac { (k+(r-1)) (k+(r-2)) (k+(r-3)) ...(k+1) (k) }
       { (r) (r-1) (r-2) .... (2)(1)} \cdot (-1)^r\\
= \frac {(k+(r-1))!}{r! (k-1)!}  \cdot (-1)^r 
= \binom{(k+(r-1)}{r} \cdot (-1)^r   
$

#### 最後確認一下【和為1】的機率分布特性

$
\sum_{r=0}^{-\infty}p_{R_k}(r)\\
= \sum_{r=0}^{\infty} \binom{-k}{r} p^k (-q)^{r}\\
= p^k \cdot \sum_{r=0}^{-\infty} \binom{-k}{r}  (-q)^{r}\\
= p^k \cdot (1-q)^{-k}\\
= p^k \cdot (p)^{-k} \\
= 1
$

## 二項分布 vs 負二項分布


### 二項分布

- $n$ (戰之數): 已知
- $K$ (勝之數): 未知
- $R$ (敗之數)= n - $K$

$
1= (p+q)^n \\
= \sum_{k=0}^{n} p^k q^r \binom{n}{k,r} \\
= \sum_{k=0}^{n} p^k q^{(n-k)} \binom{n}{k,n-k}
$

### 負二項分布

- $k$ (勝之數): 已知
- $R$ (敗之數): 未知
- $N$ (戰之數)= k + $R$

$
1=({p\over{1-q}})^k \\
= p^k \cdot (1-q)^{-k} \\
= \sum_{r=0}^{\infty} p^k q^r \binom{r+k-1}{r, k-1}\\
= \sum_{n=k}^{\infty} p^k q^{(n-k)} \binom{n-1}{n-k, k-1}
$
